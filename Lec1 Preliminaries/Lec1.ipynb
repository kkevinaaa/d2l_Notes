```python
import os

os.makedirs(os.path.join('..', 'data'), exist_ok=True)
data_file = os.path.join('..', 'data', 'house_tiny.csv')
with open(data_file, 'w') as f:
    f.write('NumRooms,Alley,Price\n')  # 列名
    f.write('NA,Pave,127500\n')  # 每行表示一个数据样本
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')

import pandas as pd
data = pd.read_csv(data_file)
print(data)
```

```python
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2]
inputs = inputs.fillna(inputs.mean(numeric_only=True))
print(inputs)
```

```python
inputs_1 = pd.get_dummies(inputs)
inputs_2 = pd.get_dummies(inputs, dummy_na=True)
# dummy_na=True 会为缺失值创建一个新的类别
print(inputs_1)
print(inputs_2)
```

```python
# 1. 刚出生的张量，内存完美连续
x = torch.arange(12)  # 形状是 [12]
# 连续状态下：view 和 reshape 都能用，且都共享内存
y_view = x.view(3, 4)
y_reshape = x.reshape(3, 4)

# ----------------------------

# 2. 我们做个转置操作（.t()），此时张量在内存里变得【不连续】了
x_transposed = y_view.t()  # 形状是 [4, 3]

# 💥 不连续状态下调用 view —— 直接崩溃报错！
# RuntimeError: view size is not compatible with input tensor's size and stride
# y_err = x_transposed.view(2, 6) 

# ✅ 不连续状态下调用 reshape —— 完美运行，全自动在后台复制了内存
y_ok = x_transposed.reshape(2, 6)
print(y_ok)
```
